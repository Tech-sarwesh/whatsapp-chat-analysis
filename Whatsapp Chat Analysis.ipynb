```python
import re
import pandas as pd
```

```python
f = open("BBIT.txt",'r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
#11/09/2024, 8:08â€¯pm - Subhrodeep Bbit Ai: This message was deleted
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APap][Mm]\s-\s'
```

```python
messages = re.split(pattern,data)[2:]
```

```python
dates = re.findall(pattern,data)[1:]
```

```python
dates
```

```python
df = pd.DataFrame({"user_message":messages,"message_date":dates})
```

```python
df
```

```python
df['message_date'] = df['message_date'].str.replace(' -', '', regex=False).str.strip()
```

```python
df[['message_date']]
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %I:%M %p').dt.strftime('%Y-%m-%d %H:%M')
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%Y-%m-%d %H:%M',errors='coerce')
```

```python
df
```

```python
df
```

```python
df.rename(columns={'message_date':'date'},inplace = True)
```

```python
df
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group notofication')
        messages.append(entry[0])
        
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace = True)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
